```python
# Install the *lab* version of neurogym (gyyang repo)
!pip install gym
!git clone https://github.com/gyyang/neurogym.git
%cd neurogym/
!pip install -e .

import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim
import time
import neurogym as ngym
import math
import random
from torch.nn import functional as F

def set_seed(seed):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)

SEED = 0
set_seed(SEED)

task_name = 'ContextDecisionMaking-v0'

ctx_timing = {
    'fixation': 300,
    'stimulus': 750,
    'delay':    1000,  # fixed delay instead of TruncExp
    'decision': 100
}

kwargs = {'dt': 20, 'timing': ctx_timing}

# choose seq_len based on trial duration
trial_ms = sum(ctx_timing.values())
dt = kwargs['dt']
trial_steps = math.ceil(trial_ms / dt)
print("Approx steps per trial:", trial_steps)

# safety margin so a sequence comfortably covers a trial
seq_len = trial_steps + 20
print("Using seq_len =", seq_len)

batch_size = 16

def make_dataset_and_env(seed):
    set_seed(seed)
    ds = ngym.Dataset(task_name, env_kwargs=kwargs,
                      batch_size=batch_size, seq_len=seq_len)
    env = ds.env
    try:
        env.seed(seed)
    except Exception:
        pass
    return ds, env

baseline_dataset, env = make_dataset_and_env(SEED)
l2_dataset, _        = make_dataset_and_env(SEED)

input_size = env.observation_space.shape[0]
output_size = env.action_space.n

print("Input size:", input_size)
print("Output size:", output_size)

inputs_np, targets_np = baseline_dataset()
print('Inputs shape (SeqLen, Batch, Dim):', inputs_np.shape)
print('Targets shape (SeqLen, Batch):', targets_np.shape)


# Baseline: Leaky RNN cell
class LeakyRNN(nn.Module):

    def __init__(self, input_size, hidden_size, dt=None, **kwargs):
        super().__init__()
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.tau = 100  # ms

        if dt is None:
            alpha = 1.0
        else:
            alpha = dt / self.tau
        self.alpha = alpha

        self.input2h = nn.Linear(input_size, hidden_size)
        self.h2h = nn.Linear(hidden_size, hidden_size)

    def init_hidden(self, input_shape):
        batch_size = input_shape[1]
        return torch.zeros(batch_size, self.hidden_size)

    def recurrence(self, input, hidden):

        h_new = torch.relu(self.input2h(input) + self.h2h(hidden))
        h_new = hidden * (1 - self.alpha) + h_new * self.alpha
        return h_new

    def forward(self, input, hidden=None):

        if hidden is None:
            hidden = self.init_hidden(input.shape).to(input.device)

        outputs = []
        steps = range(input.size(0))
        for t in steps:
            hidden = self.recurrence(input[t], hidden)
            outputs.append(hidden)

        outputs = torch.stack(outputs, dim=0)  # (seq_len, batch, hidden_size)
        return outputs, hidden

# RNNNet wrapper
class RNNNet(nn.Module):

    def __init__(self, input_size, hidden_size, output_size, **kwargs):
        super().__init__()
        self.rnn = LeakyRNN(input_size, hidden_size, **kwargs)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):

        rnn_output, _ = self.rnn(x)
        out = self.fc(rnn_output)
        return out, rnn_output

def train_model(net, dataset, n_steps=5000, print_every=100,
                beta_L2=0.0, model_name="model"):
    optimizer = optim.Adam(net.parameters(), lr=1e-3)
    criterion = nn.CrossEntropyLoss()

    running_loss = 0.0
    start_time = time.time()

    loss_history = []
    step_history = []

    print(f"Training {model_name} on {task_name}...")
    for i in range(n_steps):

        inputs_np, targets_np = dataset()
        inputs  = torch.from_numpy(inputs_np).float()
        targets = torch.from_numpy(targets_np).long()

        optimizer.zero_grad()

        outputs, rnn_activity = net(inputs)

        mask = (targets > 0)

        if mask.sum() == 0:
            continue

        outputs_dec = outputs[mask]
        labels_dec  = targets[mask]

        loss_task = criterion(outputs_dec, labels_dec)

        # Optional L2 penalty on hidden rates
        if beta_L2 > 0.0:
            rate_l2 = (rnn_activity ** 2).sum()
            loss = loss_task + beta_L2 * rate_l2
        else:
            loss = loss_task

        loss.backward()
        optimizer.step()

        loss_value = loss.item()
        loss_history.append(loss_value)
        step_history.append(i + 1)

        running_loss += loss_value
        if (i + 1) % print_every == 0:
            avg_loss = running_loss / print_every
            elapsed = time.time() - start_time
            print(f"[{model_name}] step {i+1:4d} | loss {avg_loss:.4f} | time {elapsed:.1f}s")
            running_loss = 0.0

    return net, step_history, loss_history

hidden_size = 128

set_seed(SEED)
baseline_net = RNNNet(input_size=input_size,
                      hidden_size=hidden_size,
                      output_size=output_size,
                      dt=env.dt)
print("Baseline net:\n", baseline_net)

baseline_net, base_steps, base_losses = train_model(
    baseline_net, baseline_dataset,
    n_steps=5000,
    print_every=100,
    beta_L2=0.0,
    model_name="Baseline LeakyRNN"
)

beta_rate_L2 = 9e-7
set_seed(SEED)
l2_net = RNNNet(input_size=input_size,
                hidden_size=hidden_size,
                output_size=output_size,
                dt=env.dt)
print("L2-regularised net:\n", l2_net)

l2_net, l2_steps, l2_losses = train_model(
    l2_net, l2_dataset,
    n_steps=5000,
    print_every=100,
    beta_L2=beta_rate_L2,
    model_name=f"LeakyRNN + L2_rates (β={beta_rate_L2})"
)

def test_model(net, env, num_trials=200):
    """Test a trained RNNNet on individual trials."""
    net.eval()

    trial_infos = []
    activity_dict = {}

    with torch.no_grad():
        for i in range(num_trials):
            env.new_trial()
            ob, gt = env.ob, env.gt
            T = ob.shape[0]

            inputs = torch.from_numpy(ob[:, np.newaxis, :]).float()

            outputs, rnn_activity = net(inputs)
            outputs_np = outputs.numpy()
            rnn_np = rnn_activity.numpy()

            logits_last = outputs_np[-1, 0, :]
            choice = logits_last.argmax()
            correct = (choice == gt[-1])

            trial_infos.append({
                'trial_index': i,
                'correct': bool(correct),
                'choice': int(choice),
                'gt_last': int(gt[-1]),
            })
            activity_dict[i] = rnn_np[:, 0, :]

    net.train()
    accuracy = np.mean([ti['correct'] for ti in trial_infos])
    return accuracy, trial_infos, activity_dict

def smooth(y, window=200):
    y = np.array(y)
    if len(y) < window:
        return y
    kernel = np.ones(window) / window
    return np.convolve(y, kernel, mode='valid')

base_smooth = smooth(base_losses, window=200)
l2_smooth   = smooth(l2_losses,   window=200)

plt.figure(figsize=(8, 5))

plt.plot(base_steps[:len(base_smooth)],
         base_smooth,
         linewidth=2,
         label='Baseline (smoothed)')

plt.plot(l2_steps[:len(l2_smooth)],
         l2_smooth,
         linewidth=2,
         label='L2 (smoothed)')

all_smooth = np.concatenate([base_smooth, l2_smooth])
y_min = all_smooth.min()
y_max = all_smooth.max()
margin = 0.1 * (y_max - y_min + 1e-8)
plt.ylim(y_min - margin, y_max + margin)

plt.xlabel("Training step")
plt.ylabel("Loss")
plt.title("Training curves: Baseline vs L2-regularised LeakyRNN (CDM task)")
plt.legend()
plt.tight_layout()
plt.show()


baseline_acc, baseline_trials, baseline_activity = test_model(baseline_net, env)
print("Task 2 (CTX) Baseline trial accuracy:", baseline_acc)

l2_acc, l2_trials, l2_activity = test_model(l2_net, env)
print("Task 2 (CTX) L2-regularised trial accuracy:", l2_acc)
```

```python
def loss_decomposition_avg(net, dataset, beta_L2, name="model", n_batches=200):
    criterion = nn.CrossEntropyLoss()

    task_losses = []
    reg_terms   = []

    with torch.no_grad():
        for _ in range(n_batches):
            inputs_np, targets_np = dataset()
            inputs  = torch.from_numpy(inputs_np).float()
            targets = torch.from_numpy(targets_np).long()

            outputs, rnn_activity = net(inputs)
            mask = (targets > 0)
            if mask.sum() == 0:
                continue

            outputs_dec = outputs[mask]
            labels_dec  = targets[mask]

            loss_task = criterion(outputs_dec, labels_dec)
            rate_l2   = (rnn_activity**2).sum()
            loss_reg  = beta_L2 * rate_l2

            task_losses.append(float(loss_task))
            reg_terms.append(float(loss_reg))

    mean_task = np.mean(task_losses)
    mean_reg  = np.mean(reg_terms)
    print(f"[{name}] mean task loss over {len(task_losses)} batches: {mean_task:.4f}")
    print(f"[{name}] mean reg term  over {len(reg_terms)} batches: {mean_reg:.4f}")
    print(f"[{name}] reg / task ratio: {mean_reg / mean_task:.4f}")
    print()

beta_L2 = 9e-7
loss_decomposition_avg(baseline_net, baseline_dataset, 0.0,      "Baseline")
loss_decomposition_avg(l2_net,       l2_dataset,       beta_L2, "L2 (β=1e-6)")
```

```python

import numpy as np
import torch
import matplotlib.pyplot as plt

# Ground truth labels at final timestep (one per trial)
baseline_gt = np.array([info["gt_last"] for info in baseline_trials])
l2_gt       = np.array([info["gt_last"] for info in l2_trials])

baseline_correct = np.array([info["correct"] for info in baseline_trials])
l2_correct       = np.array([info["correct"] for info in l2_trials])

print("Unique baseline labels:", np.unique(baseline_gt))
print("Unique L2 labels:",       np.unique(l2_gt))

# From ContextDecisionMaking source:
# action 0 = fixation, 1 = choice1, 2 = choice2
choice1_label = 1
choice2_label = 2

# Indices of correct choice1 / choice2 trials
baseline_choice1_trials = np.where((baseline_gt == choice1_label) & baseline_correct)[0]
baseline_choice2_trials = np.where((baseline_gt == choice2_label) & baseline_correct)[0]
l2_choice1_trials       = np.where((l2_gt == choice1_label)       & l2_correct)[0]
l2_choice2_trials       = np.where((l2_gt == choice2_label)       & l2_correct)[0]

print("Baseline correct choice1 trials:", len(baseline_choice1_trials))
print("Baseline correct choice2 trials:", len(baseline_choice2_trials))
print("L2 correct choice1 trials:",       len(l2_choice1_trials))
print("L2 correct choice2 trials:",       len(l2_choice2_trials))

# all trials by label, ignoring correctness
baseline_choice1_trials_all = np.where(baseline_gt == choice1_label)[0]
baseline_choice2_trials_all = np.where(baseline_gt == choice2_label)[0]
l2_choice1_trials_all       = np.where(l2_gt == choice1_label)[0]
l2_choice2_trials_all       = np.where(l2_gt == choice2_label)[0]

def to_numpy(arr):
    if isinstance(arr, torch.Tensor):
        return arr.detach().cpu().numpy()
    return np.asarray(arr)

def stack_activity_dict(activity_dict):
    keys = sorted(activity_dict.keys())
    trial_arrays = [to_numpy(activity_dict[k]) for k in keys]
    data = np.stack(trial_arrays, axis=1)
    return data

baseline_act = stack_activity_dict(baseline_activity)
l2_act       = stack_activity_dict(l2_activity)

T, n_trials, hidden_size = baseline_act.shape
print("Activity shape (T, trials, H):", baseline_act.shape)

dt_ms  = getattr(env, "dt", 20.0)
time_s = np.arange(T) * dt_ms / 1000.0
```

```python
# Define epoch windows using env.timing and env.dt

dt_ms = getattr(env, "dt", 100.0)

timing = env.unwrapped.timing

def steps(period_name):

    return int(timing[period_name] / dt_ms)

fix_T      = steps('fixation')
stim_T     = steps('stimulus')
delay_T    = steps('delay')
decision_T = steps('decision')

fix_start      = 0
stim_start     = fix_start    + fix_T
delay_start    = stim_start   + stim_T
decision_start = delay_start  + delay_T

stim_win     = slice(stim_start,     stim_start     + stim_T)
delay_win    = slice(delay_start,    delay_start    + delay_T)
decision_win = slice(decision_start, decision_start + decision_T)

def epoch_mean(activity, trial_idx, epoch_slice):
    if len(trial_idx) == 0:
        return np.zeros(activity.shape[2])
    epoch_act = activity[epoch_slice, :, :][:, trial_idx, :]
    return epoch_act.mean(axis=(0, 1))

baseline_corr_trials = np.where(baseline_correct)[0]
l2_corr_trials       = np.where(l2_correct)[0]

baseline_stim_mean = epoch_mean(baseline_act, baseline_corr_trials, stim_win)
l2_stim_mean       = epoch_mean(l2_act,       l2_corr_trials,       stim_win)

baseline_delay_mean = epoch_mean(baseline_act, baseline_corr_trials, delay_win)
l2_delay_mean       = epoch_mean(l2_act,       l2_corr_trials,       delay_win)

baseline_dec_mean = epoch_mean(baseline_act, baseline_corr_trials, decision_win)
l2_dec_mean       = epoch_mean(l2_act,       l2_corr_trials,       decision_win)

# Plot histograms for all three epochs
fig, axes = plt.subplots(1, 3, figsize=(12, 4),
                         sharey=True, sharex=True)


for ax, base, l2_vals, title in zip(
    axes,
    [baseline_stim_mean, baseline_delay_mean, baseline_dec_mean],
    [l2_stim_mean,       l2_delay_mean,       l2_dec_mean],
    ["Stimulus", "Delay", "Decision"],
):
    ax.hist(base,   bins=30, alpha=0.5, label="Baseline")
    ax.hist(l2_vals, bins=30, alpha=0.5, label="L2")
    ax.set_title(title)
    ax.set_xlabel("Mean hidden activity")
    ax.set_ylabel("Number of neurons")

axes[0].legend()
plt.tight_layout()
plt.show()
```

```python
# compute a common range across all epochs + models
global_max = max(
    baseline_stim_mean.max(), l2_stim_mean.max(),
    baseline_delay_mean.max(), l2_delay_mean.max(),
    baseline_dec_mean.max(),   l2_dec_mean.max()
)
bin_edges = np.linspace(0, global_max, 31)

fig, axes = plt.subplots(1, 3, figsize=(12, 4), sharey=True, sharex=True)

for ax, base, l2_vals, title in zip(
    axes,
    [baseline_stim_mean, baseline_delay_mean, baseline_dec_mean],
    [l2_stim_mean,       l2_delay_mean,       l2_dec_mean],
    ["Stimulus", "Delay", "Decision"],
):
    ax.hist(base,    bins=bin_edges, alpha=0.5, label="Baseline")
    ax.hist(l2_vals, bins=bin_edges, alpha=0.5, label="L2")
    ax.set_title(title, fontsize=12)
    ax.set_xlabel("Mean hidden activity", fontsize=11)
    ax.set_ylabel("Number of neurons", fontsize=11)
    ax.tick_params(axis='both', labelsize=9)

axes[0].legend(fontsize=9)
plt.tight_layout()
plt.show()
```

```python
# Population mean activity over time

T, n_trials, n_hidden = baseline_act.shape

def population_mean_timecourse(activity, correct_mask):
    corr_idx = np.where(correct_mask)[0]
    if len(corr_idx) == 0:
        return np.zeros(activity.shape[0])
    sub = activity[:, corr_idx, :]
    return sub.mean(axis=(1, 2))

baseline_mean_t = population_mean_timecourse(baseline_act, baseline_correct)
l2_mean_t       = population_mean_timecourse(l2_act,       l2_correct)

time_s = np.arange(T) * (dt_ms / 1000.0)

plt.figure(figsize=(7, 4))
plt.plot(time_s, baseline_mean_t, label="Baseline", linewidth=2)
plt.plot(time_s, l2_mean_t,       label="L2",       linewidth=2)

epoch_boundaries = {
    "stimulus": stim_start,
    "delay": delay_start,
    "decision": decision_start,
}
for name, idx in epoch_boundaries.items():
    t_b = idx * dt_ms / 1000.0
    plt.axvline(t_b, linestyle="--", alpha=0.3)

trial_end = (decision_start + decision_T) * dt_ms / 1000.0
plt.axvline(trial_end, linestyle="--", alpha=0.3)

plt.xlabel("Time (s)", fontsize=12)
plt.ylabel("Mean hidden activity (population)", fontsize=12)
plt.title("Population activity over time, Baseline vs L2 (CDM task)", fontsize=13)

plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend(fontsize=10)

plt.tight_layout()
plt.show()
```
